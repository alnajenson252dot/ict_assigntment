PART 2 5TH QUESTION

```python
total_medals_per_country = medal_winners['NOC'].value_counts().reset_index()
total_medals_per_country.columns = ['NOC', 'Total Medals']
display(total_medals_per_country.head())
```

```python
display(total_medals_per_country.head(11))
```

6TH QUESTION

```python
import matplotlib.pyplot as plt
import seaborn as sns
top_10_countries = total_medals_per_country.head(10)
sns.barplot(x='NOC', y='Total Medals', data=top_10_countries)
plt.title('Top 10 Countries by Total Medals')
plt.xlabel('Country (NOC)')
plt.ylabel('Total Medals')
plt.show()
```

7TH QUESTION

```python
gold_medals = medal_winners[medal_winners['Medal'] == 'Gold']
print(f"Number of Gold Medals: {len(gold_medals)}")
display(gold_medals.head())
```

```python
gold_medals_per_country = gold_medals['NOC'].value_counts().reset_index()
gold_medals_per_country.columns = ['NOC', 'Gold Medals']
display(gold_medals_per_country.head(10))
```

```python
top_10_gold_countries = gold_medals_per_country.head(10)
sns.barplot(x='NOC', y='Gold Medals', data=top_10_gold_countries)
plt.title('Top 10 Countries by Gold Medals')
plt.xlabel('Country (NOC)')
plt.ylabel('Gold Medals')
plt.show()
```

```python
medals_by_gender = medal_winners['Sex'].value_counts().reset_index()
medals_by_gender.columns = ['Sex', 'Total Medals']
display(medals_by_gender)
```

9TH QUESTION

```python
gender_trend_over_years = medal_winners.groupby(['Year', 'Sex']).size().reset_index(name='Total Medals')
display(gender_trend_over_years.head())
```

```python
sns.lineplot(x='Year', y='Total Medals', hue='Sex', data=gender_trend_over_years, marker='o')
plt.title('Medal Trend Over Years by Gender')
plt.xlabel('Year')
plt.ylabel('Total Medals')
plt.show()
```
